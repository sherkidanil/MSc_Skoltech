```python
from __future__ import annotations

import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def logistic_map(r: float, n_iter: int, x0: float,) -> np.array:
    x = np.zeros(n_iter)
    x[0] = x0

    for i in range(1, n_iter):
        x[i] = r * x[i-1] * (1 - x[i-1])

    return x
```

```python
res = logistic_map(3, 500, 0.6)
plt.figure(figsize=(8,4), dpi=300)
plt.plot(np.arange(0, res.size), res)
plt.grid(True)
plt.xlabel('iter')
plt.ylabel('x')
plt.title(fr'Logistic map for $r$={3} and $x_0$={0.6}')
plt.xlim(0)
plt.ylim(0)
plt.show()
```

```python
def create_bifurcation(n: int, m: int, r_list: np.array, x0: float) -> None:
    plt.figure(figsize=(12,6), dpi=300)

    for r in r_list:
        if x0 is not None:
            log_map = logistic_map(r = r, n_iter = n+m, x0 = x0)
        else:
            log_map = logistic_map(r = r, n_iter = n+m, x0 = random.random())

        plt.scatter([r]*m, log_map[n:], c='tab:blue', s=1, alpha=.5)

    plt.grid(True)
    plt.xlabel('r')
    plt.ylabel('x')
    plt.title('Bifurcation map')
    plt.xlim(0)
    plt.ylim(0)
    plt.show()
```

```python
create_bifurcation(n = 200, m = 200, r_list = np.linspace(0,4,500), x0 = 0.4)
```
