# Question 1 (4 points)

and $r > 0$ are parameters.


$$ \min_{x \in \mathbb{R}^n: \Vert x-c \Vert \leq r } \Vert z -x \Vert_2$$

## Solution

Let's rewrite this problem by squaring it

$$ \min_{x \in \mathbb{R}^n: \Vert x - c \Vert^2 \leq r^2 } \Vert z -x \Vert^2_2$$

Thus, we have only one inequality constraint and zero equality, according to general problem

Lagrangian $L$ equal

$$ L(x, \lambda) =  \Vert z -x \Vert^2_2 + \lambda \cdot (\Vert x - c \Vert^2 - r^2)  $$

Dual function

$$ g(\lambda) = \inf_x (L(x, \lambda)) = \inf_x (\Vert z -x \Vert^2_2 + \lambda \cdot (\Vert x - c \Vert^2 - r^2))$$

Let's find optimal value of $L$

For this task let's rewrite $L$ as $L = (z-x, z-x) + \lambda (x-c, x - c) - \lambda r^2$

$$ L = (z-x)^{\top} (z - x) + (x-c)^{\top} (x - c)  - \lambda r^2 $$

$$  = -2z + 2x + \lambda x - \lambda c = (1 + \lambda) x - (z + \lambda c)$$

$$ \nabla_x L = 0: (1 + \lambda) x - (z + \lambda c) $$ 

$$ (1 + \lambda) x = z + \lambda c $$

$$ x^{*} = \frac{z + \lambda c}{1 + \lambda} $$

## Solution

Let's square $ f(x)$, because $\inf_x \Vert x-z \Vert_2 =\inf_x \Vert x-z \Vert^2_2 $

$$ L(x, v) = \Vert x - z \Vert^2_2 - v ((a,x) - b)  = \Vert x - z \Vert^2_2 - v (a,x) + v b $$

$$ \nabla_x L = 2 (x-z) -v a$$

$$ \nabla_x L = 0: 2x - 2z - va = 0 $$

$$ x^* = z + \frac{v}{2} a $$

$$ g(v) = L(x^*, v) = \Vert  z + \frac{v}{2}  - z \Vert^2_2 - v (a, z + \frac{v}{2} a) + v b = 
\frac{v^2}{4} \Vert  a \Vert^2_2 - (v a, z + \frac{v}{2} a) + v b = \frac{v^2}{4} \Vert  a \Vert^2_2 - (v a, z) -\frac{v^2}{2}(a,  a) + v b $$

$$ v^* = \frac{2((a,z) - b)}{\Vert a \Vert^2}$$

# Question 6 (4 points)
Solve the problem using the gradient with projection method:

$$\min (c,x)$$
s.t. $(a_i,x) \leq b_i$,
where $A \in \mathbb{R}^{(n+1) \times n},b \in \mathbb{R}^{n+1}, x \in \mathbb{R}^n$, with $i = 1,\dots,101; n = 100$.

$$A = \begin{bmatrix}
-1  & 0 & 0 & \dots & 0 \\
0  & -1 & 0 & \dots & 0 \\
\ldots  & \ldots & \ldots & \ldots & \dots \\
0  & 0 & 0 & \dots & -1 \\
1  & 1 & 1 & \dots & 1
\end{bmatrix}, 
b = [0,0,0,\dots,1]^{\top}, c = [1,1,1, \dots, 1]^{\top}.$$

```python
import numpy as np
import pandas as pd

n = 100
x = np.ones((n,1))*3
b = np.zeros((n+1,1))
b[-1] = 1

A = np.array(np.random.rand(n,n))
A = np.zeros_like(A) - 1
A = np.diag(np.diag(A))
A_app = np.zeros_like(x)+1
A = np.concatenate((A, np.reshape(A_app,(1,100))), axis=0)
c = np.ones((n,1))
```

```python
def task6_func(x):
    return float(c.T @ x)

def proj(x):
    for i in range(len(b)):
        if float(A[i].T @ x)>b[i]:
            x = x - (float(-b[i]+A[i]@x)/np.linalg.norm(A[i]))*np.reshape(A[i],(n,1))
        else:
            pass
    return x

def pdg(x0, gamma, tol, func = task6_func, proj_func = proj, grad = c):
    x = x0
    x_prev = np.zeros_like(x)
    k = 0

    x_arr = [x]
    f_arr = [func(x)]
    k_arr = [k]

    while np.linalg.norm(x - x_prev) > tol:
        x_prev = x
        k += 1
        x = proj_func(x - gamma*grad)

        x_arr.append(x)
        f_arr.append(func(x))
        k_arr.append(k)
    
    df = pd.DataFrame({'step': k_arr, 'f': f_arr})
    return df, x_arr
```

```python
df, x_arr = pdg(x, 1e-3, 1e-4)
print(fr'The optimal value $ f^*$ = {df.f[3]}, was found with {df.step[3]} iterations, and the optimal vector $x^* = ${x_arr[-1].T}')
```
