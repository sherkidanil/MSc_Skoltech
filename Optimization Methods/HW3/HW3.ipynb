# Question 1 (1 point)

Prove that if functions $f(.),g(.)$ are convex, then $h(x) = \max{f(x),g(x)} $ is convex as well.

## Solution 

Let $a,b \in \mathbb{R},x_1,x_2 \in \mathbb{R}$ such that $a+b=1$ and $x=ax_1+bx_2$.

Since $𝑓(.)$ and $𝑔(.)$ are convex, we have:

$$f(ax_1+bx_2) \leq af(x_1)+bf(x_2)$$

and

$$g(ax_1+bx_2) \leq ag(x_1)+bg(x_2)$$

Therefore, 

$$h(ax_1+bx_2)=\max(f(ax_1+bx_2),g(ax_1+bx_2))$$

$$ \leq \max(af(x_1)+bf(x_2),  ag(x_1)+bg(x_2))$$

$$\leq  a \max(f(x_1),   g(x_1)) +  b \max(f(x_2),  g(x_2))$$

$$ \leq ah(x_1) +  b h(x_2)$$

Therefore, $h(.)$ is convex.

**Alternative solution**

Let $x_1, x_2 \in \mathbb{R}$ and $t \in [0, 1]$. 

We have to prove that $h( t x_1 + (1-t)x_2) \leq t h(x_1) + (1-t)h(x_2)$(*)

$$f(t x_1 + (1-t)x_2) \leq t f(x_1) + (1-t)f(x_2)$$

$$g(t x_1 + (1-t)x_2) \leq t g(x_1) + (1-t)g(x_2)$$

case, when $f(x)\geq g(x)$: $h(x)=f(x)$

$$h( t x_1 + (1-t)x_2) = f(t x_1 + (1-t)x_2) \leq t f(x_1) + (1-t)f(x_2) \leq t h(x_1) + (1-t)h(x_2)$$

case, when $g(x)> f(x)$: $h(x)=g(x)$

$$h( t x_1 + (1-t)x_2) = g(t x_1 + (1-t)x_2) \leq t g(x_1) + (1-t)g(x_2) \leq t h(x_1) + (1-t)h(x_2)$$

Therefore, in box case condition (*) True, $h(x)$ is convex.

# Question 2 (2 points)

Find conjugate function $f^∗(y)$ for $f(x) = (Qx,x)$ and $Q ≻ 0.$

## Solution

1. **$x^*$ finding**

By definition,

$$ f^* (y) = \sup_x \{(y,x) - f(x)\}$$

Let's find extremum

$$f^*_x{'} = \frac{\partial f^*}{\partial x} = y^{\top} - (Q^{\top} + Q)x$$

Extremum point: $f^*_x{'} = 0$

$$y^{\top} - (Q^{\top} + Q)x^* = 0$$

$$x^* = y^{\top}(Q^{\top} + Q)^{-1} = y^{\top} (2Q)^{-1} $$

2. **$f^*$ finding**

$$ f^* (y) = \sup_x \{(y,x^*) - f(x^*)\}$$ 

$$f^* (y) = (y, x^*) - (Qx^*, x^*)$$

$$ f^* (y) = (y, y^{\top} (2Q)^{-1}) - (Qy^{\top} (2Q)^{-1}, y^{\top} (2Q)^{-1}) $$

$$ f^* (y) = (y, (2Q)^{-1} y) - (Q(2Q)^{-1} y, (2Q)^{-1} y)$$

$$(2Q)^{-1} = \frac{1}{2}Q^{-1}$$

$Q$ is PSD by definition, so $(Q^{-1})^{\top} = Q^{-1}$

$$ f^* (y) = \frac{1}{2} y^{\top}Q^{-1} y  - \frac{1}{2}y^{\top} Q^{-1} Q Q^{-1} y $$

$$ f^* (y) = \frac{1}{2} y^{\top}Q^{-1} y  - \frac{1}{2}y^{\top} Q^{-1} Q \frac{1}{2} Q^{-1} y $$

$$ f^* (y) = \frac{1}{2} y^{\top}Q^{-1} y  - \frac{1}{2}y^{\top} Q^{-1} Q \frac{1}{2} Q^{-1} y $$

$$f^* (y) = \frac{1}{2} y^{\top}Q^{-1} y - \frac{1}{4}y^{\top} Q^{-1} y$$

$$f^* (y) = \frac{1}{4}y^{\top} Q^{-1} y$$

## Initialization and scpectrum analyzing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


#Hilbert matrix 
n = 5
A = [[1.0/(i+j-1) for i in range(1,n+1)] for j in range(1,n+1)] 
A = np.array(A)

def func(x: np.array):
    return x.T @ A @ x

def gradient(x: np.array):
    return (A+A.T) @ x

# hessian
H = A+A.T
```

## Methods implementation and calculating

```python
def gradient_descent(x0, gamma, tol, func=func, grad_func=gradient,
                         adapt_gamma=False, gamma_max=1, c1 = 0.01, gamma_step = 0.01):
    x = x0
    x_prev = np.zeros_like(x)
    k = 0
    k1 = 0

    x_arr = [x]
    f_norm = [0 - np.sqrt(np.power(func(x),2))]
    x_norm = [np.linalg.norm(np.array([0,0,0,0,0]) - x, 2)]
    f_arr = [func(x)]
    k_arr = [k]
    k1_arr  = [k1]
    gamma_arr = [gamma]

    func_calls_f, grad_calls_f, hess_calls_f = 0, 0, 0
    func_calls_x, grad_calls_x, hess_calls_x = 0, 0, 0
    k_f, k_x = 0, 0

    while np.linalg.norm(grad_func(x), 2) > tol:
        x_prev = x
        if adapt_gamma:
            k1 = 0
            gamma = gamma_max
            grad_val = grad_func(x)
            while func(x - gamma*grad_val)>(func(x) - c1 * gamma * np.dot(grad_val.T, grad_val)):
                gamma -= gamma_step
                k1 += 1
            k1_arr.append(k1)
            gamma_arr.append(gamma)
        k += 1
        x = x - gamma*grad_func(x)

        x_arr.append(x)
        x_norm.append(np.linalg.norm(np.array([0,0,0,0,0]) - x))
        f_norm.append(np.sqrt(np.power(0 - func(x),2)))
        f_arr.append(func(x))
        k_arr.append(k)

        if (func(x) - 0 < 1e-5) and (grad_calls_f == 0):
            if adapt_gamma:
                func_calls_f = k*2
                grad_calls_f = k*2
                hess_calls_f = 0
                k_f = k
            else:
                func_calls_f = 0
                grad_calls_f = k*1
                hess_calls_f = 0
                k_f = k

        if (np.linalg.norm(x,2) < 1e-5) and (grad_calls_x == 0):
            if adapt_gamma:
                func_calls_x = k*2
                grad_calls_x = k*2
                hess_calls_x = 0
                k_x = k
            else:
                func_calls_x = 0
                grad_calls_x = k*1
                hess_calls_x = 0
                k_x = k
        
    if adapt_gamma:
        df = pd.DataFrame({'step': k_arr, 'f': f_arr, 'gamma': gamma_arr,
                           'x_norm': x_norm,'f_norm': f_norm, 'k1': k1_arr})
    else:
        df = pd.DataFrame({'step': k_arr, 'f': f_arr, 'x_norm': x_norm,'f_norm': f_norm})

    calls = [[func_calls_f, func_calls_x],
             [grad_calls_f, grad_calls_x],
             [hess_calls_f, hess_calls_x],
             [k_f, k_x]]

    return df, x_arr, calls
```

```python
import warnings
warnings.filterwarnings('ignore')

def results_report(df, x_arr, calls, i = 0, j = 20, slice_plot=True):
    display(df)
    print('x* = ',x_arr[-1])
    plt.figure(figsize = (6,3))
    plt.plot(df.step, np.log10(df.x_norm))
    plt.ylabel(r'$\log_{10} \Vert x^* - x \Vert_2$')
    plt.xlabel('Iteration step')
    plt.title('x convergence for all iteration process')
    plt.axhline(y = -5, ls = '--', c='r')
    plt.grid(True)
    plt.show()

    plt.figure(figsize = (6,3))
    plt.plot(df.step, np.log10(df.f_norm))
    plt.ylabel(r'$\log_{10} \Vert f^* - f \Vert_2$')
    plt.xlabel('Iteration step')
    plt.title('f convergence for all iteration process')
    plt.grid(True)
    plt.axhline(y = -5, ls = '--', c='r')
    plt.show()

    if slice_plot:
        plt.figure(figsize = (6,3))
        plt.plot(df.step[i:j], np.log10(df.x_norm[i:j]),'-*')
        plt.ylabel(r'$\log_{10} \Vert x^* - x \Vert_2$')
        plt.xlabel('Iteration step')
        plt.title(f'x convergence for {i}:{j} iterations')
        plt.grid(True)
        plt.show()

        plt.figure(figsize = (6,3))
        plt.plot(df.step[i:j], np.log10(df.f_norm[i:j]), '-*')
        plt.ylabel(r'$\log_{10} \Vert f^* - f \Vert_2$')
        plt.xlabel('Iteration step')
        plt.title(f'f convergence for {i}:{j} iterations')
        plt.grid(True)
        plt.show()

    df_calls = pd.DataFrame(calls, columns = ['|f-f*|<1e-5','|| x-x*||<1e-5'],
                            index = ['Func calls', 'Grad calls', 'Hess_calls', 'Iterations'])
    def q_calc(df):
        q_arr = []
        for i in range(1,len(df.f)):
            q_arr.append((df.f[i] - 0)/(df.f[i-1] - 0))
        plt.plot(range(len(q_arr)),q_arr)
        plt.grid(True)
        plt.xlabel('step')
        plt.ylabel('Convergence rate q')
        plt.xscale('log')
        plt.show()

    q_calc(df)

    display(df_calls)
```

### Case 3. Exact line search

```python
def gradient_descent_exact_line_search(x0, tol, func=func, grad_func=gradient):
    x = x0
    x_prev = np.zeros_like(x)
    k = 0

    x_arr = [x]
    f_norm = [0 - np.sqrt(np.power(func(x),2))]
    x_norm = [np.linalg.norm(np.array([0,0,0,0,0]) - x, 2)]
    f_arr = [func(x)]
    k_arr = [k]

    func_calls_f, grad_calls_f, hess_calls_f = 0, 0, 0
    func_calls_x, grad_calls_x, hess_calls_x = 0, 0, 0
    k_x, k_f = 0, 0

    while np.linalg.norm(grad_func(x), 2) > tol:
        x_prev = x
        gamma = (grad_func(x).T @ grad_func(x))/(grad_func(x).T @ (H @ grad_func(x)))
        k += 1
        x = x - gamma*grad_func(x)

        x_arr.append(x)
        x_norm.append(np.linalg.norm(np.array([0,0,0,0,0]) - x))
        f_norm.append(np.sqrt(np.power(0 - func(x),2)))
        f_arr.append(func(x))
        k_arr.append(k)

        if (func(x) - 0 < 1e-5) and (grad_calls_f == 0):
            func_calls_f = 0
            grad_calls_f = k*1
            hess_calls_f = k*1
            k_f = k

        if (np.linalg.norm(x,2) < 1e-5) and (grad_calls_x == 0):
            func_calls_x = 0
            grad_calls_x = k*1
            hess_calls_x = k*1
            k_x = k
    
    calls = [[func_calls_f, func_calls_x],
             [grad_calls_f, grad_calls_x],
             [hess_calls_f, hess_calls_x],
             [k_f, k_x]]

    df = pd.DataFrame({'step': k_arr, 'f': f_arr, 'x_norm': x_norm,'f_norm': f_norm})
    return df, x_arr, calls
```

In this case also we have linear convergence in one of the three segments, where convergence rate **q** equal to **0.95**

### Case 4. Heavy ball method

```python
def heavy_ball_method(x0, tol, func=func, grad_func=gradient):
    x = x0
    x_prev = np.zeros_like(x)
    k = 0
    k1 = 0

    func_calls_f, grad_calls_f, hess_calls_f = 0, 0, 0
    func_calls_x, grad_calls_x, hess_calls_x = 0, 0, 0
    k_x, k_f = 0, 0

    # IMPORTANT PART
    alpha = 4./np.power((np.sqrt(L) + np.sqrt(m)),2)
    beta = np.power((np.sqrt(L) - np.sqrt(m))/(np.sqrt(L) + np.sqrt(m)),2)

    x_arr = [x]
    f_norm = [0 - np.sqrt(np.power(func(x),2))]
    x_norm = [np.linalg.norm(np.array([0,0,0,0,0]) - x, 2)]
    f_arr = [func(x)]
    k_arr = [k]
    k1_arr  = [k1]
    gamma_arr = [gamma]

    while np.linalg.norm(grad_func(x), 2) > tol:
        x_prev = x

        k += 1
        if k>=2:
            x = x - alpha*grad_func(x) + beta*(x - x_arr[-2])
        else:
            x = x - alpha*grad_func(x) + beta*(x - x_arr[-1])

        if (func(x) - 0 < 1e-5) and (grad_calls_f == 0):
            func_calls_f = 0
            grad_calls_f = k*1
            hess_calls_f = k*0
            k_f = k

        if (np.linalg.norm(x,2) < 1e-5) and (grad_calls_x == 0):
            func_calls_x = 0
            grad_calls_x = k*1
            hess_calls_x = k*0
            k_x = k


        x_arr.append(x)
        x_norm.append(np.linalg.norm(np.array([0,0,0,0,0]) - x))
        f_norm.append(np.sqrt(np.power(0 - func(x),2)))
        f_arr.append(func(x))
        k_arr.append(k)
            
    calls = [[func_calls_f, func_calls_x],
             [grad_calls_f, grad_calls_x],
             [hess_calls_f, hess_calls_x],
             [k_f, k_x]]
    df = pd.DataFrame({'step': k_arr, 'f': f_arr, 'x_norm': x_norm,'f_norm': f_norm})
    return df, x_arr, calls
```
